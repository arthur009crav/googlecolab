```python
import matplotlib.pyplot as plt
#importando a biblioteca
```

```python
anos = [1950,1960,1970,1980,1990,2000,2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

plt.plot(anos,gdp, color='green',marker = 'o', linestyle ='dashed' ,linewidth=2, markersize=9)


plt.title('Analise de Dados !')

plt.ylabel('Bilhões')
plt.xlabel('Anos')

plt.show()

#gráfico simples , leia a documentação , gráfico simples, que vc pode mexer em diversas coisas como rotulo, eixo, 
#estilos de linha marcadores, e etc, Verifique a documentação ! 

```
